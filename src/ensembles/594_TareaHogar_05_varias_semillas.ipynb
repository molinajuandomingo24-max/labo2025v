Aqui se definen los hiperparámetros de LightGBM que participan de la Bayesian Optimization
<br> si es un numero entero debe ir  makeIntegerParam
<br> si es un numero real (con decimales) debe ir  makeNumericParam
<br> es muy importante leer cuales son un lower y upper  permitidos y ademas razonables

```r


# Definición de SOLO los 3 hiperparámetros principales a optimizar
PARAM$hypeparametertuning$hs <- makeParamSet(
  makeIntegerParam("num_leaves", lower= 16L, upper= 1024L), # Rango ajustado
  makeIntegerParam("min_data_in_leaf", lower= 1000L, upper= 8000L), # Rango ajustado
  makeNumericParam("feature_fraction", lower= 0.3, upper= 1.0) # Rango ajustado
)
```

2.2.5 Configuracion Bayesian Optimization

```r
# La función EstimarGanancia_AUC_lightgbm se mantiene IGUAL
EstimarGanancia_AUC_lightgbm <- function(x) {

  # Obtener la semilla de la corrida actual
  semilla_actual <- PARAM$semillas_bo[PARAM$indice_corrida]

  # x pisa (o agrega) a param_fijos
  param_completo <- modifyList(PARAM$lgbm$param_fijos, x)
  param_completo$seed <- semilla_actual # Asignar la semilla

  # entreno LightGBM
  modelocv <- lgb.cv(
    data= dtrain,
    nfold= PARAM$hyperparametertuning$xval_folds,
    stratified= TRUE,
    param= param_completo
  )

  # obtengo la ganancia
  AUC <- modelocv$best_score

  # hago espacio en la memoria
  rm(modelocv)
  gc(full= TRUE, verbose= FALSE)

  message(format(Sys.time(), "%a %b %d %X %Y"), " AUC ", AUC, " (Semilla: ", semilla_actual, ")")

  return(AUC)
}
```

```r
# Aquí comienza la configuración de la Bayesian Optimization

funcion_optimizar <- EstimarGanancia_AUC_lightgbm
configureMlr(show.learner.output= FALSE)

obj.fun <- makeSingleObjectiveFunction(
  fn= funcion_optimizar,
  minimize= FALSE,
  noisy= TRUE,
  par.set= PARAM$hypeparametertuning$hs,
  has.simple.signature= FALSE
)

# Definir el control MBO fuera del loop
kbayesiana_base <- "bayesiana"
ctrl <- makeMBOControl(save.on.disk.at.time= 600)
ctrl <- setMBOControlTermination(ctrl, iters= PARAM$hyperparametertuning$iteraciones)
ctrl <- setMBOControlInfill(ctrl, crit= makeMBOInfillCritEI())

surr.km <- makeLearner(
  "regr.km",
  predict.type= "se",
  covtype= "matern3_2",
  control= list(trace= TRUE)
)

# Inicializo variables para rastrear el mejor resultado global
best_global_AUC <- -1
best_global_hyperparams <- NULL
resultados_semillas <- data.table()

# Loop para correr la BO con cada semilla
for (s in 1:length(PARAM$semillas_bo)) {

  PARAM$indice_corrida <- s
  semilla_actual <- PARAM$semillas_bo[s]
  kbayesiana <- paste0(kbayesiana_base, "_s", semilla_actual, ".RDATA")

  message(format(Sys.time(), "%a %b %d %X %Y"), " --- INICIANDO CORRIDA CON SEMILLA: ", semilla_actual, " ---")

  # Establecer la semilla para mlrMBO
  set.seed(semilla_actual)
  ctrl$save.file.path <- kbayesiana # Actualizar el path de guardado

  # inicio la optimizacion bayesiana
  bayesiana_salida <- mbo(obj.fun, learner= surr.km, control= ctrl)

  # Proceso el resultado de la corrida actual
  tb_bayesiana_actual <- as.data.table(bayesiana_salida$opt.path)
  setorder(tb_bayesiana_actual, -y)

  # Registro el mejor AUC de esta corrida
  mejor_auc_corrida <- tb_bayesiana_actual[1, y]

  # Guardo el resultado para el análisis global
  tb_bayesiana_actual[, iter := .I]
  tb_bayesiana_actual[, semilla := semilla_actual]
  resultados_semillas <- rbind(resultados_semillas, tb_bayesiana_actual, fill=TRUE)

  # Verifico si es el mejor resultado global
  if (mejor_auc_corrida > best_global_AUC) {
    best_global_AUC <- mejor_auc_corrida
    best_global_hyperparams <- tb_bayesiana_actual[
      1,
      setdiff(colnames(tb_bayesiana_actual),
        c("y","dob","eol","error.message","exec.time","ei","error.model",
          "train.time","prop.type","propose.time","se","mean","iter", "semilla")),
      with= FALSE
    ]
  }
}

# Almaceno los resultados finales de las 5 corridas

# Guardo el log completo de todas las semillas
fwrite(resultados_semillas,
  file= "BO_log_multi_semilla.txt",
  sep= "\t"
)

# Los mejores hiperparámetros globales
PARAM$out$lgbm$mejores_hiperparametros <- best_global_hyperparams
PARAM$out$lgbm$y_global <- best_global_AUC

write_yaml( PARAM, file="PARAM.yml")

print("Mejores Hiperparámetros Globales (entre 5 semillas):")
print(PARAM$out$lgbm$mejores_hiperparametros)
print(paste("Mejor AUC Global:", PARAM$out$lgbm$y_global))
```
